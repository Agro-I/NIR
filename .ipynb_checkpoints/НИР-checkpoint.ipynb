```python
def load_image(name, output=False):
    img = cv2.imread(path + '/' + name + '.png')
    return img

def load_mask(name, output=False):
    img_path = path + '/' + name + 'm.png'
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    return np.array(img).reshape(img.shape[0], img.shape[1], 1)


def display(display_list, title_list=['Input Image', 'True Mask', 'Predicted Mask']):
    plt.figure(figsize=(15, 15))

    for i in range(len(display_list)):
        plt.subplot(1, len(display_list), i + 1)
        plt.title(title_list[i])
        plt.imshow(display_list[i])
        plt.axis('off')
    plt.show()
```

```python
def print_cluster(result, cluster_tag):
    result_copy = np.zeros_like(result)
    slice_img = np.where(result == cluster_tag, 1, 0)
    result_copy[..., np.newaxis] = slice_img[..., np.newaxis]
    return result_copy


def get_IoU_for_slice(slice_, mask):
    intersection = np.logical_and(slice_, mask)
    union = np.logical_or(slice_, mask)
    iou_score = np.sum(intersection) / np.sum(union)
    return iou_score


def get_IoU(slices):
    return np.array(slices).max(), np.argmax(np.array(slices))

def accuracy_IoU_scorer(result, mask):
    metrics_iou = []
    for i in np.unique(result.flatten()):
        result_copy = result.copy()
        result_copy[abs(result_copy ) ==i] = 1
#         result_copy[result_copy != 1] = 0
        metrics_iou.append(get_IoU_for_slice(result_copy, mask))

    accuracy, cluster_ind = get_IoU(metrics_iou)
    print('---', accuracy,cluster_ind, np.unique(result.flatten())[cluster_ind])
    slice_ = print_cluster(result, np.unique(result.flatten())[cluster_ind])
    return accuracy, slice_
```

```python
def calculate_IoU(predicted_mask, true_mask):
    unique_clusters = np.unique(predicted_mask.flatten())
    best_IoU = 0
    best_cluster_label = None

    for cluster in unique_clusters:
        # Бинаризация предсказанной маски для текущего кластера
        predicted_cluster_mask = np.where(predicted_mask == cluster, True, False)

        # Пересечение и объединение масок
#         intersection = np.logical_and(predicted_cluster_mask, true_mask)
#         union = np.logical_or(predicted_cluster_mask, true_mask)

        # Вычисление метрики IoU
#         IoU = np.sum(intersection) / np.sum(union)
        intersection = predicted_cluster_mask*true_mask # Logical AND
        union = predicted_cluster_mask + true_mask # Logical OR
        IoU = intersection.sum()/float(union.sum())
        # Обновление лучшего значения IoU и метки кластера
#         print(IoU)
        if IoU > best_IoU:
            best_IoU = IoU
            best_cluster_label = cluster
    labels = np.where(predicted_mask == best_cluster_label, best_cluster_label, 0)

    return best_IoU, labels

# Пример использования
# predicted_mask = np.array([[ 1, 1, 1],
#                             [ 0, 1, 1],
#                             [ 0, 0, 0]])

# true_mask = np.array([[ 0, 1, 1],
#                       [ 0, 1, 1],
#                       [ 0, 0, 0]])

# best_IoU, best_cluster_label = calculate_IoU(predicted_mask, true_mask)
# print("Best IoU:", best_IoU)
# print("Best Cluster Label:", best_cluster_label)
```

```python
def blur_img(img, blur_param=30, center=(100, 110), axes=(100, 60), angle=0,startAngle = 0, endAngle = 360):
    blurred_img = cv2.GaussianBlur(img, (21, 21), blur_param)
    mask = np.zeros(img.shape, dtype=np.uint8)
    mask = cv2.ellipse(mask, center=center, axes=axes, angle=angle,startAngle=startAngle,endAngle=endAngle, color=(255, 255, 255), thickness=-1)
    out = np.where(mask == np.array([255, 255, 255]), img, blurred_img)
    return out
```

```python
def dbscan(data, eps=0.01, min_samples=6, metric='euclidean'):
    print(f"Начало выполнения DBSCAN с параметрами eps={eps}, min_samples={min_samples}")
    
    db = DBSCAN(eps=eps, min_samples=min_samples, metric=metric, n_jobs=-1).fit(data)
    labels = db.labels_

    n_clusters = len(np.unique(labels)) - (1 if -1 in labels else 0)
    n_noise = list(labels).count(-1)

#     print("Оцененное количество кластеров: %d" % n_clusters)
#     print("Оцененное количество выбросов: %d" % n_noise)

    labeled_img = labels.reshape(200, 200)

    return labeled_img,db.labels_
```

```python
def meanShift(data,bandwidth,cluster_all):
    print(f"Начало выполнения MeanShift с параметрами bandwidth={bandwidth}, cluster_all={cluster_all}")
    meanshift = MeanShift(bandwidth=bandwidth,cluster_all=True).fit(data)
    labels = meanshift.labels_
    n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
    n_noise = list(labels).count(-1)

    print("Оцененное количество кластеров: %d" % n_clusters)
    print("Оцененное количество выбросов: %d" % n_noise)

    labeled_img = labels.reshape(200,200)
    return labeled_img
```

```python
def k_means(data,n_clusters,algorithm,max_iter):
    print(f"Начало выполнения k_means с параметрами n_clusters={n_clusters}, algorithm={algorithm}")
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, algorithm=algorithm, max_iter=max_iter).fit(data)
    labels = kmeans.labels_
    n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
    n_noise = list(labels).count(-1)

    # print("Оцененное количество кластеров: %d" % n_clusters)
    # print("Оцененное количество выбросов: %d" % n_noise)

    labeled_img = labels.reshape(200,200)
    return labeled_img
```

```python
def objective(trial):
    n_clusters = trial.suggest_int('n_clusters', 2, 10)
    max_iter = trial.suggest_int('max_iter', 1000, 2000)
    algorithm=trial.suggest_categorical('algorithm', [ "elkan", "auto", "full"])

    db = k_means(img_3f,n_clusters=n_clusters, algorithm=algorithm, max_iter=max_iter)

    # Calculate accuracy using accuracy_IoU_scorer function
    accuracy, _ = accuracy_IoU_scorer(db, mask)
    
    return accuracy

for i in range(n):
    img_3f = blur_img(images[i].reshape((-1, 3)))
    mask=masks[i]
    study = optuna.create_study(direction='maximize')
    study.optimize(objective, n_trials=100)
    best_trial = study.best_trial
    n_clusters = best_trial.params['n_clusters']
    algorithm = best_trial.params['algorithm']
    max_iter = best_trial.params['max_iter']
    bestParamDBscan = k_means(img_3f,n_clusters=n_clusters, algorithm=algorithm,max_iter=max_iter)
    accuracy, slice_ = accuracy_IoU_scorer(bestParamDBscan, masks[i])
    print(accuracy)
    display([blur_img(images[i]),mask,slice_])
    directory = 'out/images/imagesKMeans/'
    if not os.path.exists(directory):
        os.makedirs(directory)
    with open('out/images/imagesKMeans.txt', "a") as f:
        f.write(uniq_img_names[i] + ' ' + str(accuracy) + ' ' + str(n_clusters)+ ' ' + str(algorithm) + '\n')
        f.close()

    data_2d = slice_.reshape((-1, slice_.shape[-1]))
    np.savetxt('out/images/imagesKMeans/' + (uniq_img_names[i]) + '.txt', data_2d, delimiter=',')
```
